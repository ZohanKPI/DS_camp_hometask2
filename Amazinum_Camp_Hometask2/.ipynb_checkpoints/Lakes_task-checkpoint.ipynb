# The task:
Generate 20 random variable from 1 to 100.

Draw the plot of the sequence.

Lets assume each point represent the height and so all plotting is 2d mountains.

Then consider the unlimited rain from above - cavities become lakes full of water.

Determine the deepest lake.

```python
import numpy as np
import matplotlib.pyplot as plt

def find_deepest_lake(terrain):
    deepest_point = 101  # Initialize the deep point with a value greater than what will be found
    deepest_point_index = 0  # Initialize the deepest point index

    for i in range(1, len(terrain) - 1):  # Iterate through the indices of the terrain array
        if (terrain[i - 1] > terrain[i]) and (terrain[i + 1] > terrain[i]) and (terrain[i] < deepest_point):
            # Check if the current element is a deep point and update the value and index of the deep point
            deepest_point = terrain[i]
            deepest_point_index = i

    left_part = terrain[:deepest_point_index]  # Define the left part of the array up to the deep point
    right_part = terrain[(deepest_point_index + 1):]  # Define the right part of the array after the deep point

    max_left = np.max(left_part)  # Find the maximum in the left part
    max_right = np.max(right_part)  # Find the maximum in the right part

    max_left_indices = np.where(terrain == np.max(left_part))[0]  # Find the indices of maximum values in the left part
    max_right_indices = np.where(terrain == np.max(right_part))[0]  # Find the indices of maximum values in the right part

    max_left_index = max_left_indices[max_left_indices < deepest_point_index][-1]
    # Find the index of the maximum value in the left part that is less than the deep point index

    max_right_index = max_right_indices[max_right_indices > deepest_point_index][0]
    # Find the index of the maximum value in the right part that is greater than the deep point index

    if max_right > max_left:  # Check which side has a larger maximum
        left_bottom = max_left_index
        for i in range(deepest_point_index, max_right_index+1):
            if terrain[i] - max_left >= 0:
                right_bottom = i
                break
    elif max_right < max_left:  # Check which side has a smaller maximum
        right_bottom = max_right_index
        for i in range(deepest_point_index, max_left_index-1, -1):
            if terrain[i] - max_right >= 0:
                left_bottom = i
                break
    else:  # If maximums on both sides are equal
        left_bottom = max_left_index
        right_bottom = max_right_index + deepest_point_index - 2

    return left_bottom, right_bottom
```

```python
def visualize_terrain_and_lake(terrain):
    start, end = find_deepest_lake(terrain)  
    if start is None:
        print("No lake found!")  
        plt.plot(terrain, label='Terrain')  
    else:
        lake = terrain[start:end+1]
        plt.plot(terrain, label='Terrain', color='gray')
        plt.plot(range(start, end+1), lake, color='red', label='lake')
    plt.title('Terrain and Lake Visualization')
    plt.xlabel('Index')
    plt.ylabel('Height')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
np.random.seed(2023)
my_arr = np.random.randint(1, 100, 20)
print(my_arr)
visualize_terrain_and_lake(my_arr)
```
